```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on = 'Country Code', how='left')
```

```python
final_df.head()
```

```python
## to check data types we can use
final_df.dtypes
```

```python
final_df.columns
```

```python
Country_names=final_df.Country.value_counts().index
Country_names
```

```python
Country_values=final_df.Country.value_counts().values
Country_values
```

```python
##creating piechart for top  countries

plt.pie(Country_values[:3],labels=Country_names[:3],autopct='%1.2f%%')
```

##Observations
1. maximum records for zomato are from india after that united states and united kingdom

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

 ## conclusions/observations
1. whenever rating is between 4.5 to 4.9 this indicates food delivered is excellent
2. when rating is between 4.0 to 4.4 food delivery is good
3. if ratimg is between 3.5 to 3.9 rating is good
4. if rating is between 3.0 to 3.4 is average
5. if rating is between 2.5 to 2.9 is also average
6. if rating is between 2.9 to 2.4 is also poor
7. 

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data= ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data= ratings,palette=['blue','Red','yellow','green'])
```

## observations

1. not rated count is very high
2. maximum numbers of rating are between are between 2.5 to 3.4

```python
# count plot

sns.countplot(x='Rating color',data=ratings,palette=['blue','Red','yellow','green'])
```
